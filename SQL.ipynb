```python
# -- Problem Statement: ratio of male and female is 68:32. The store owner want to increase the number of female customers
# -- Which strategy you should offer her?

With total as (
SELECT
	Gender,
	count(`Customer ID`) as client_count

FROM customer_shopping_trends
Group by Gender
Order by client_count desc
)
Select
	sum(case when Gender = 'Male' then client_count else 0 end) count_male,
	sum(case when Gender = 'Female' then client_count else 0 end) count_Female,
	round(sum(case when Gender = 'Male' then client_count else 0 end)/
     (sum(case when Gender = 'Male' then client_count else 0 end)
     + sum(case when Gender = 'Female' then client_count else 0 end)) * 100,2) pct_male,

	round(sum(case when Gender = 'Female' then client_count else 0 end)/
     (sum(case when Gender = 'Male' then client_count else 0 end)
     + sum(case when Gender = 'Female' then client_count else 0 end)) * 100,2) pct_female
from total
;
```

```python
# -- Female Age Group Distribution

SELECT

	Case
     when Age between 18 and 31 then '18-31'
     when Age between 32 and 45 then '32-45'
     when Age between 46 and 59 then '46-59'
     when Age between 60 and 70 then '60-70'
	end as age_groups,
    count(`Customer ID`) as female_customer


FROM customer_shopping_trends
Where gender = 'Female'
Group by age_groups
Order by Female_Customer desc;
```

```python
# -- Category distribution by Female Age Group

SELECT
	Category,
    count(`Customer ID`) as purchased_amount
FROM customer_shopping_trends
Where gender = 'Female'
Group by Category
Order by purchased_amount desc;
```

```python
# -- `Item Purchased` distribution by Female Age Group

SELECT
	`Item Purchased`,
    count(`Customer ID`) as purchased_number
FROM customer_shopping_trends
Where gender = 'Female'
Group by `Item Purchased`
Order by purchased_number desc;
```

```python
# -- Distribution by season

SELECT
	Season,
    count(`Customer ID`) as purchased_number
FROM customer_shopping_trends
Where gender = 'Female'
Group by Season
Order by purchased_number desc;
```

```python
# -- Payment methods preference

SELECT
	`Payment Method`,
    count(`Customer ID`) as count_method
FROM customer_shopping_trends
Where gender = 'Female'
Group by `Payment Method`
Order by count_method desc;
```

```python
# -- Shipping types preference

SELECT
	`Shipping Type`,
    count(`Customer ID`) as count_type
FROM customer_shopping_trends
Where gender = 'Female'
Group by `Shipping Type`
Order by count_type desc;
```
